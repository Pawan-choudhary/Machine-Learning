The **make_pipeline()** function is a utility function provided by the scikit-learn library in Python. It is used to create a pipeline object by sequentially concatenating multiple transformers and an estimator. The resulting pipeline can be used to streamline the process of transforming data and fitting a model.

```python
from sklearn.pipeline import make_pipeline

pipeline = make_pipeline(PolynomialFeatures(1), StandardScaler(), LinearRegression())
pipeline.fit(X_train, y_train)

print(pipeline.score(X_train,y_train))
print(pipeline.score(X_test,y_test))
```

# Cross-validation Types:
Cross-validation is a resampling technique used in machine learning to assess the performance of a model and estimate its generalization ability. It involves partitioning the available data into subsets to train and validate the model on different combinations of these subsets. Here are some commonly used types of cross-validation:

* K-Fold Cross-Validation: In k-fold cross-validation, the data is divided into k equally sized folds or subsets. The model is trained and evaluated k times, each time using a different fold as the validation set and the remaining folds as the training set. The performance results are then averaged to obtain an overall assessment of the model's performance. K-fold cross-validation helps in reducing the variance of the estimated performance compared to a single validation set.

* Stratified K-Fold Cross-Validation: Stratified k-fold cross-validation is an extension of k-fold cross-validation that ensures the class distribution is maintained across the folds. It is particularly useful when dealing with imbalanced datasets, where the distribution of classes is uneven. Stratified k-fold cross-validation helps ensure that each fold represents the class distribution of the overall dataset, providing a more reliable evaluation.

* Leave-One-Out Cross-Validation (LOOCV): In leave-one-out cross-validation, each data point is used as a validation set, and the model is trained on the remaining data points. This process is repeated for each data point, resulting in n iterations for a dataset with n samples. LOOCV provides an unbiased estimate of model performance but can be computationally expensive for large datasets.

* Leave-P-Out Cross-Validation (LPOCV): Leave-p-out cross-validation is a generalization of LOOCV, where p data points are left out as the validation set, and the model is trained on the remaining data. This process is repeated for all possible combinations of p data points, resulting in multiple iterations. LPOCV allows for a more flexible trade-off between computation time and evaluation accuracy compared to LOOCV.

* Time Series Cross-Validation: Time series cross-validation is specifically designed for evaluating models on time series data, where the temporal ordering of data points is important. It involves splitting the data into training and validation sets based on a chronological order, with earlier data used for training and later data used for validation. Time series cross-validation helps assess the model's ability to generalize to future unseen data.

```python
from sklearn.model_selection import KFold
#from sklearn.preprocessing import make_pipeline

kf = KFold(n_splits=10)

degrees = 32
train_scores = []
test_scores = []

for degree in range(1,degrees):

  fold_train_scores = []
  fold_test_scores = []
    
  for train_index, test_index in kf.split(X):
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    pipeline = make_pipeline(PolynomialFeatures(degree), scaler, LinearRegression())
    pipeline.fit(X_train, y_train)
    train_score = pipeline.score(X_train, y_train)
    test_score  = pipeline.score(X_test,  y_test)
    fold_train_scores.append(train_score)
    fold_test_scores.append(test_score)
  train_score = np.mean(fold_train_scores)
  test_score = np.mean(fold_test_scores)
  train_scores.append(train_score)
  test_scores.append(test_score)

plt.figure()
plt.plot(list(range(1,32)), train_scores, label='train')
plt.plot(list(range(1,32)),test_scores, label='Validation Average Score')
plt.legend(loc='lower right')  
plt.xlabel("Degree")
plt.ylabel("R-score")
plt.ylim(0,1)
plt.show()
```

# Regularization

```python
from sklearn.linear_model import  Ridge, Lasso, ElasticNet

np.random.seed(1)

X = np.random.rand(100,1) # generates a NumPy array X with dimensions 50x1, containing random values between 0 and 1

y = 2.7*(X**5) - 2.1*(X**4) + 2.3*(X**3) + 0.2*(X**2) + 0.3*X + 0.4*np.random.rand(100,1)  

degree = 4
poly = PolynomialFeatures(degree)
x_poly = poly.fit_transform(X)
x_poly_scaled = scaler.fit_transform(x_poly)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_poly_scaled,y, test_size=0.2, random_state=1)

linear = LinearRegression()
linear.fit(X_train, y_train)
print("Linear Model Coef")
print(linear.coef_)

# L2
ridge = Ridge(alpha=1)  # Shrikage parameter
ridge.fit(X_train, y_train)
print("Ridge Regression Coef")
print(ridge.coef_)

# L1
lasso = Lasso(alpha=1) 
lasso.fit(X_train, y_train)
print("lasso Regression Coef")
print(lasso.coef_)

# L1 + L2
elastic_net = ElasticNet(alpha=1)
elastic_net.fit(X_train, y_train)
print("elasticnet Regression Coef")
print(elastic_net.coef_)

print("-"*30)

print("Train Performance")
print(" R-square for Linear Regression :", linear.score(X_train, y_train))
print(" R-square for Lasso Regression :", lasso.score(X_train, y_train))
print(" R-square for Ridge Regression :", ridge.score(X_train, y_train))
print(" R-square for ElasticNet Regression :", elastic_net.score(X_train, y_train))

print("-"*30)

print("Test Performance")
print(" R-square for Linear Regression :", linear.score(X_test, y_test))
print(" R-square for Lasso Regression :", lasso.score(X_test, y_test))
print(" R-square for Ridge Regression :", ridge.score(X_test, y_test))
print(" R-square for ElasticNet Regression :", elastic_net.score(X_test, y_test))


```
